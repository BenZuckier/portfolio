## use model

### Imports

```python
import joblib
import pandas as pd
import numpy as np
import sklearn.preprocessing as sk_prep
from sklearn.model_selection import train_test_split
import xgboost as xgb
from sklearn.metrics import accuracy_score, recall_score, roc_auc_score, precision_score
```

### Load models from file

```python
onehot_model = joblib.load('/content/models/onehot')
xgb_model = joblib.load('/content/models/xgb')
toOnehot = joblib.load('/content/models/labelCols')
colNames = joblib.load('/content/models/onehottedLabels')
```

### evaluate performance

```python
# should prob make this stuff into a self contained module
X_test = pd.read_csv('/content/train_transaction.csv.zip', compression='zip').drop("TransactionID", axis=1)
v310 = X_test[["V310"]].copy()
v307 = X_test[["V307"]].copy()
X_test = X_test.loc[:,~X_test.columns.str.startswith('V')]
X_test["V310"] = v310[["V310"]]
X_test["V307"] = v307[["V307"]]
X_test['TimeInDay'] = X_test.TransactionDT % 86400
X_test['Cents'] = X_test.TransactionAmt % 1


X_test_enc = X_test.sample(frac=0.35)
y_test = X_test_enc[["isFraud"]].copy()
X_test_enc = X_test_enc.drop("isFraud",axis=1)

onehotted_test = onehot_model.transform(X_test_enc[toOnehot])
onehotDat_test =  pd.DataFrame(onehotted_test, columns=colNames, index=X_test_enc.index)
X_test_enc = pd.concat([X_test_enc.drop(toOnehot, axis=1), onehotDat_test], axis=1)

pred = xgb_model.predict(X_test_enc)
pred_proba = xgb_model.predict_proba(X_test_enc)
xgb_df = pd.DataFrame(data=[accuracy_score(y_test, pred), recall_score(y_test, pred),
                precision_score(y_test, pred), roc_auc_score(y_test, pred_proba[:,1])], 
            columns=['XGB Score'],
            index=["Accuracy", "Recall", "Precision", "ROC AUC Score"])
xgb_df
```

### understand the model?

```python
from xgboost import plot_importance
import matplotlib.pyplot as plt

%matplotlib inline
plt.rcParams['figure.figsize'] = [30, 15]
plot_importance(xgb_model, max_num_features=30)
plt.show()
```

```python
from sklearn.metrics import plot_precision_recall_curve, precision_recall_curve, plot_roc_curve, roc_curve, plot_confusion_matrix
from sklearn.metrics import confusion_matrix, accuracy_score, roc_auc_score, classification_report, recall_score, f1_score

plt.rcParams['figure.figsize'] = [10, 10]
plot_confusion_matrix(xgb_model, X_test_enc, y_test, values_format = '', display_labels=['Not Fraud', 'Fraud'])
plt.grid(None)
print(classification_report(y_test, pred, zero_division=0))
```

```python
from sklearn import metrics

auc = metrics.roc_auc_score(y_test, pred_proba[:,1])

false_positive_rate, true_positive_rate, thresolds = metrics.roc_curve(y_test, pred_proba[:,1])

plt.rcParams['figure.figsize'] = [25, 15]
plt.figure(figsize=(8, 8), dpi=100)
plt.axis('scaled')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.title("AUC & ROC Curve")
plt.plot(false_positive_rate, true_positive_rate, 'g')
plt.fill_between(false_positive_rate, true_positive_rate, facecolor='lightgreen', alpha=0.7)
plt.text(0.95, 0.05, 'AUC = %0.4f' % auc, ha='right', fontsize=12, weight='bold', color='blue')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.show()
```
